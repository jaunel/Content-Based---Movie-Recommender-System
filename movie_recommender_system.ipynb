```python
#Importing the libraries
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

**Data Pre-Processing**

```python
#merging the dataset to one
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
credits.shape
```

```python
# we will keep only required features in our dataset rest will be neglected
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
#checking for null values
movies.head()
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
#Looking for duplicated row
movies.duplicated().sum()
```

```python
#Now based upon overview, genre, keyword, cast & crew we will create taglines

movies['genres'][0]
```

```python
#We want only values of the key (name)
def convert(list):
  l=[]
  for i in ast.literal_eval(list):       #for converting string of list to list only
    value = i['name']
    l.append(value)
  return l
movies['genres'] = movies['genres'].apply(convert)
```

```python
#for converting string of list to list
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies.head()
```

```python
# We will apply same function to keywords feature also
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords'][0]
```

```python
movies.head()
```

```python
#Now we will take only first 3 actors name from the cast
def convert2(obj):
  j=0
  list=[]
  for i in ast.literal_eval(obj):
    if j<3:
      list.append(i['name'])
      j+=1
  return list  
```

```python
movies['cast'] = movies['cast'].apply(convert2) 
```

```python
movies.head()
```

```python
   movies['crew'][0]
```

```python
# we want only director name from the crew
def fetch_director(list):
  l=[]
  for i in ast.literal_eval(list):
    if i['job'] == 'Director':
      value = i['name']
      l.append(value)
      break
  return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# Now we convert overview into list from string
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda y:y.split(" "))
```

```python
movies.head(1)
```

```python
# We want here to form one word for each entity so that we got unique tags
# To iterate over each element of the list we will use list comprehension
movies['genres'] = movies['genres'].apply(lambda x: [y.replace(" ","") for y in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [y.replace(" ","") for y in x])
movies['cast'] = movies['cast'].apply(lambda x: [y.replace(" ","") for y in x])
movies['crew'] = movies['crew'].apply(lambda x: [y.replace(" ","") for y in x])
```

```python
movies.head()
```

```python
# concatenation of all features into one
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
#creating new dataset for our problem
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
#converting list into para
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
#Now our dataset is ready
```

**Text Vectorization**

```python
#Now we will convert tags into vectors using Bag of words technique
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
# Now we will apply stemming to reduce the complexity
import nltk 
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)  
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
#cosine similarity will tell us similarity of each movie w.r to other movie
```

```python
cosine_similarity(vectors).shape
```

```python
#This will sort all movie similar to the given movie in an numpy array(we took top 5 from that)
sorted(list(enumerate(similarity[0])),reverse = True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]   #will provide index of given movie
  distances = similarity[movie_index]                       #finding movie vector
  movies_list = sorted(list(enumerate(similarity[movie_index])),reverse = True,key=lambda x:x[1])[1:6]  #selecting top 5 movie

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```
